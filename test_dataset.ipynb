```python
import numpy as np
import matplotlib.pyplot as plt
from utils.data import CustomDataset
from torch.utils.data import DataLoader
```

```python
dataset = CustomDataset("train", augmentation=True)
len(dataset)
```

```python
dloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
frames, sensors, labels = next(iter(dloader))
```

```python
frames.shape, sensors.shape, labels.shape
```

```python
frames = (frames[0].permute(0, 2, 3, 1).numpy() + 1)/2
sensors = sensors[0].numpy()
```

```python
frames[frames > 1] = 1
frames[frames < 0] = 0
```

```python
fig, axes = plt.subplots(1, 10, figsize=(20, 2))

for i in range(10):
    axes[i].imshow(frames[i])
    axes[i].set_xticks([])
    axes[i].set_yticks([])

plt.show()
print(labels[0])
```
